```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
df = pd.read_csv("/content/Heart_new2.csv")

# Corrected column assignment based on inferred data order and types
df.columns = ['HeartDisease', 'BMI', 'Smoking', 'AlcoholDrinking', 'Stroke',
              'PhysicalHealth', 'MentalHealth', 'DiffWalking', 'Sex',
              'AgeCategory', 'Race', 'Diabetic', 'PhysicalActivity',
              'GenHealth', 'SleepTime', 'Asthma', 'KidneyDisease',
              'SkinCancer']

print(df.head())
df.info()
print(df.isnull().sum())

target = df.groupby('HeartDisease').size()
print(target)

df['HeartDisease'].value_counts().plot(kind='bar', figsize=(8,6), color=["salmon","lightblue"])
plt.title("Heart Disease Distribution")
plt.show()

hd = df.groupby('HeartDisease')['HeartDisease'].count()
plt.figure(figsize=(8,6))
plt.pie(hd, labels=['Absent','Present'], colors=['salmon','lightblue'], autopct='%0.0f%%')
plt.title('Heart Disease Population %')
plt.show()

df.Sex.value_counts()
pd.crosstab(df.HeartDisease, df.Sex)

pd.crosstab(df.HeartDisease, df.Sex).plot(kind="bar",
                                          figsize=(10,6),
                                          color=["salmon","lightblue"])
plt.title("Heart Disease Frequency For Sex")
plt.xlabel("0 = Absent, 1 = Present")
plt.ylabel("Count")
plt.legend(["Female","Male"])
plt.xticks(rotation=0)
plt.show()

plt.figure(figsize=(10,6))
sns.countplot(x='AgeCategory', hue='HeartDisease', data=df)
plt.title("Heart Disease by Age Category")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(data=df['BMI'], color='salmon')
plt.title("BMI Distribution")
plt.show()

plt.figure(figsize=(8,5))
sns.countplot(x='Smoking', hue='HeartDisease', data=df)
plt.title("Smoking vs Heart Disease")
plt.show()

plt.figure(figsize=(8,5))
sns.countplot(x='AlcoholDrinking', hue='HeartDisease', data=df)
plt.title("Alcohol Drinking vs Heart Disease")
plt.show()

plt.figure(figsize=(8,5))
sns.countplot(x='PhysicalActivity', hue='HeartDisease', data=df)
plt.title("Physical Activity vs Heart Disease")
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x='HeartDisease', y='SleepTime', data=df)
plt.title("Sleep Time vs Heart Disease")
plt.show()

plt.figure(figsize=(10,6))
sns.countplot(x='GenHealth', hue='HeartDisease', data=df)
plt.title("General Health vs Heart Disease")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10,6))
sns.countplot(x='Diabetic', hue='HeartDisease', data=df)
plt.title("Diabetes vs Heart Disease")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10,6))
sns.countplot(x='Stroke', hue='HeartDisease', data=df)
plt.title("Stroke vs Heart Disease")
plt.show()

plt.figure(figsize=(10,6))
sns.countplot(x='KidneyDisease', hue='HeartDisease', data=df)
plt.title("Kidney Disease vs Heart Disease")
plt.show()

plt.figure(figsize=(10,6))
sns.countplot(x='Asthma', hue='HeartDisease', data=df)
plt.title("Asthma vs Heart Disease")
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(x='HeartDisease', y='PhysicalHealth', data=df)
plt.title("Physical Health vs Heart Disease")
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(x='HeartDisease', y='MentalHealth', data=df)
plt.title("Mental Health vs Heart Disease")
plt.show()

corr_matrix = df.corr(numeric_only=True)

plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True, cmap="YlGnBu")
plt.title("Correlation Heatmap")
plt.show()

plt.figure(figsize=(8,6))
sns.scatterplot(x='BMI', y='SleepTime', hue='HeartDisease', data=df)
plt.title("BMI vs Sleep Time")
plt.show()

plt.figure(figsize=(8,6))
sns.scatterplot(x='PhysicalHealth', y='SleepTime', hue='HeartDisease', data=df)
plt.title("Physical Health vs Sleep Time")
plt.show()

df.to_csv("Heart_Disease_Processed.csv", index=False)
```
